```r
dml_res$pmlog10fdr <- -log10(dml_res$fdr + 1e-30)
dml_res[1:3,]
```

```r

# Get the sequence lengths for the hg19 genome
hg19_seqlengths <- seqlengths(BSgenome.Hsapiens.UCSC.hg19)

gr_seqlengths <- hg19_seqlengths[seqlevels(gr)]
# gr_seqlengths

outdir <- "out/ipynb/export_differential_to_bw"
dir.create(outdir)

for (metric in c("mu1", "mu2", "diff", "pmlog10fdr")) {
    gr <- GRanges(
        seqnames = paste0(
            "chr",
            dml_res$chr
        ),
        ranges = IRanges(
            start = dml_res$pos,
            end = dml_res$pos
        ),
        score = dml_res[[metric]]
    )
    # Assign the new seqlengths vector to 'gr'
    seqlengths(gr) <- gr_seqlengths
    # Create a BigWigFile connection
    con <- BigWigFile(
        file.path(
            outdir,
            paste0(
                metric,
                ".bw"
            )
        )
    )

    # Export the GRanges object to a bigwig file
    export(gr, con)
}

```
